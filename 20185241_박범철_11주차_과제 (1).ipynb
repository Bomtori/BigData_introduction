### **1. 다중회귀분석 (50점)**

* 아래의 코드 작성란에 코드 작성후 결과 출력하세요.

````
#다중회귀분석 시 필요한 패키지 설치
install.packages(c("tidyverse", "data.table", "caret", "e1071", "Hmisc"))
library(tidyverse)
library(data.table)
options(repr.plot.width=10, repr.plot.height=10)
````

````
#사용할 데이터셋을 불러옵니다.
df <- read.csv("/content/sample_data/insurance.csv")

#데이터셋을 확인합니다.
head(df, n = 10)

#구조를 확인합니다.
str(df)
````

````
#다중회귀분석을 수행합니다.
#조건 1. Charges를 반응변수로 사용, 나머지는 설명변수로 사용
model_1 = lm(charges ~ age+sex+bmi+children+smoker+region, data = df)
model_1
````

````
summary(model_1)
````

````
#조건2 : createDataPartition을 이용하여 Train/Test set 분리 후 회귀모델 생성.
#createDataPartition을 이용하여 Train/Test set을 분리합니다.
index <- caret::createDataPartition(y = df$charges, p = 0.8, list = FALSE)
train <- df[index, ]
test <- df[-index, ]

train %>% show()
test %>% show()
````

````
#회귀모델 생성
model_2 = lm(charges ~ age+sex+bmi+children+smoker+region, data = train)
model_2
````

- 회귀분석 결과
1. 회귀식 도출
* ```
b0 = -12038.6, b1 = 264.7, b2 = -125.1, b3 = 333.1, b4 = 325.8, b5 = 24017.8, b6 = -256.6, b7 = -814.6, b8 =  -821.1으로부터
charges =-12038.6 + 264.7 * age + -125.1 * sex + ...
``` 
2. age, sex, bmi, children, smoker, region이 각각 1 증가할 때, charges는 264.7, -125.1, 333.1, 325.8, 24017.8이 증가

````
#조건 3 : Summary 함수를 사용하여 회귀분석 결과 해석을 작성.
summary(model_2)
````

<결과의 해석>
```
1. 제일 밑에 F-statistic의 P-value 값은 2.2e -16로 0.05보다 작기에 이 회귀식은 회귀분석 모델 전체에 대해 통계적으로 의미가 있다고 볼 수 있다.

2. 중간의 Coefficients : 에는 y절편 값(Intercept) 및 변수들의 p-value값이 나와있다.
sex, region 변수의 경우 0.05보다 크기에 charges를 설명하는데 유의하지 않다.

3. 밑 부분의 Adjusted R-squared 값은 0.7501로, 75%만큼의 설명력을 가진다고 판단할 수 있다.

``` 

````
# 조건4. 분석 결과에 기반하여 Feature Selection(변수 선택) 및 새로운 회귀모델 생성
#유의하지 않은 변수를 누락하고 회귀모형을 새롭게 정의
model_3 = lm(charges ~ age+bmi+children+smoker, data = train)
model_3
````

````
summary(model_3)
````

````
#조건5 : Test set을 이용한 Predict(예측)결과 비교.
#test 데이터 확인
test %>% show()
````

````
#테스트 데이터를 활용하여 예측 시작
model_3_r <- predict(model_3, test)
test <- test %>% bind_cols(tibble(model_3_predict = model_3_r))
test
````

```
## 조건
1. Stroke Prediction 데이터를 로드 후 다음과 같은 전처리 과정을 거친다.
   - 결측값 제거(bmi 컬럼값이 N/A로 작성되어 있는 값을 결측값으로 지정함)
   - 결측값 제거 후 bmi 데이터타입 변경

   - 추출된 데이터와 stroke 값이 1인 데이터 병합
2. 병합한 Stroke Prediction 데이터를 caret 패키지의 createDataPartition을 이용하여 train / test 으로 분리할 것 
3. Train dataset을 사용하여glm 함수를 사용하여 모델을 생성하고 반응변수는 stroke로 설정하고, 설명변수는 stroke를 제외한 모든 것을 사용한다. 
   - step, update을 생성한 모델을 새롭게 정의할 것. 
   - 생성한 모델과 새롭게 정의한 모델의 summary결과를 확인한다.
4. Test set을 사용해서 모델을 예측한다. 예측 결과와 원본 데이터와 비교하여 모델의 적합성에 대하여 기술한다.

````
#조건1 : Stroke Prediction 데이터를 로드 후 다음과 같은 전처리 과정을 거친다.
#사용할 데이터셋을 불러옵니다.
df <- read.csv("/content/sample_data/Stroke Prediction Dataset.csv")

#데이터셋을 확인합니다.
head(df, n = 10)

#구조를 확인합니다.
str(df)
````

````
summary(df)
````

````
#조건 1-1 : 결측값 제거(bmi 컬럼값이 N/A로 작성되어 있는 값을 결측값으로 지정함)
df[df$bmi == "N/A",]=NA

#N/A값이 NA로 바꼈는지 확인
head(df, n= 10)
````

````
#결측값 제거
df = na.omit(df)
````

````
#결측값 제거 됐는지 확인
colSums(is.na(df))
````

````
#조건 1-2 : 결측값 제거 후 bmi 데이터타입 변경
df$bmi <- as.numeric(df$bmi)

#bmi의 데이터타입이 numeric으로 변경 됐는지 확인
str(df$bmi)
````

````
#조건 1-3 : Stroke 값이 0인 데이터를 랜덤으로 209개 추출

#stroke 값이 0인 것을 새로운 변수에 정의합니다.
df_stroke_zero <- df[df$stroke == 0, ]

df_stroke_zero
````

````
#sample함수를 사용하여 랜덤으로 209개를 추출합니다.
df_stroke_r <- df_stroke_zero[sample(nrow(df_stroke_zero), 209), ]

df_stroke_r
````

````
# 추출된 데이터와 stroke 값이 1인 데이터 병합

#stroke 값이 1인 것을 새로운 변수에 정의합니다.
df_stroke_one <- df[df$stroke == 1, ]

stroke_Prediction <- merge(df_stroke_one, df_stroke_r, all = T)

stroke_Prediction
````

### **2. 로지스틱 회귀분석 (50점)**

* 아래의 코드 작성란에 코드 작성후 결과 출력하세요.

````
#병합한 Stroke Prediction 데이터를 caret 패키지의 createDataPartition을 이용하여 train / test 으로 분리할 것 
index <- caret::createDataPartition(y = stroke_Prediction$stroke, p = 0.8, list = FALSE)
train <- stroke_Prediction[index, ]
test <- stroke_Prediction[-index, ]

train %>% show()
test %>% show()
````

````
str(train)
````

````
#Train dataset을 사용하여glm 함수를 사용하여 모델을 생성하고 반응변수는 stroke로 설정하고, 설명변수는 stroke를 제외한 모든 것을 사용한다. 

#glm(stroke ~., data = train) : stroke는 반응변수, .으로 모든 설명변수를 나타낸다.
# family = "binomial" :  종속 변수가 어떤 분포를 따르고 있는지 옵션을 주는 것
m <- glm(stroke ~., data = train, family = "binomial")
summary(m)
````

````
#step, update을 생성한 모델을 새롭게 정의할 것. 

m2 <- step(m, direction = "backward")
````

````
#생성한 모델과 새롭게 정의한 모델의 summary결과를 확인한다.

summary(m2)
````

````
#step, update을 생성한 모델을 새롭게 정의할 것. 
#생성한 모델과 새롭게 정의한 모델의 summary결과를 확인한다.

m3 <- update(m, .~ .)
summary(m3)
````

````
# Test set을 사용해서 모델을 예측한다. 예측 결과와 원본 데이터와 비교하여 모델의 적합성에 대하여 기술한다. 

#테스트 데이터 확인
test %>% show()
````

````
predict_value <- predict(m, test, type = "response") %>% tibble(predict_value = .)
predict_value %>% show()
````

````
predict_value <- predict(m2, test, type = "response") %>% tibble(predict_value = .)
predict_value %>% show()
````

````
predict_value <- predict(m3, test, type = "response") %>% tibble(predict_value = .)
predict_value %>% show()
````

````
predict_check <- test %>% select(stroke) %>% dplyr::bind_cols(., predict_value) 
predict_check %>% show()
````

````
# 예측 결과와 원본 데이터와 비교하여 모델의 적합성에 대하여 기술한다. 
predict(m2, newdata = test, type = "response") %>%
  tibble(predict_stroke = .) %>% bind_cols(test, .)
````

````
# 예측 결과와 원본 데이터와 비교하여 모델의 적합성에 대하여 기술한다. 
predict(m3, newdata = test, type = "response") %>%
  tibble(predict_stroke = .) %>% bind_cols(test, .)
````
